```python
# This magic command is used in Jupyter Notebooks to display plots inline (right below the code cell)
%matplotlib inline

# Define a function to perform Gradient Descent for Linear Regression
def gradient_descent(x, y):
    # Initialize slope (m) and intercept (b) to 0
    m_curr = b_curr = 0

    # Learning rate (controls the step size of each iteration)
    rate = 0.01

    # Total number of data points
    n = len(x)

    # Plot the original data points in red
    plt.scatter(x, y, color='red', marker='+', linewidth='5')

    # Perform 10,000 iterations to adjust m and b
    for i in range(10000):
        # Predict y using current slope and intercept
        y_predicted = m_curr * x + b_curr

        # Plot the predicted line (green) for visualization at each step (optional but good for understanding)
        plt.plot(x, y_predicted, color='green', alpha=0.01)

        # Calculate the gradient (partial derivatives) for m and b
        md = -(2/n) * sum(x * (y - y_predicted))  # derivative w.r.t. m (slope)
        yd = -(2/n) * sum(y - y_predicted)        # derivative w.r.t. b (intercept)

        # Update m and b using gradient descent formula
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

# Only use this in Jupyter Notebook to display plots inline
%matplotlib inline

# Input features (x) and target values (y)
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Define the gradient descent function for linear regression
def gradient_descent(x, y):
    m_curr = b_curr = 0          # Initialize slope and intercept
    rate = 0.01                  # Learning rate
    n = len(x)                   # Number of data points

    # Plot original data points
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='red', marker='+', linewidth=3)

    # Iterate to minimize the error
    for i in range(10000):
        y_predicted = m_curr * x + b_curr     # Predict using current m and b

        # Plot prediction lines at intervals for visualization (not all iterations)
        if i % 1000 == 0:
            plt.plot(x, y_predicted, color='green', alpha=0.5)

        # Compute gradients (partial derivatives)
        md = -(2/n) * sum(x * (y - y_predicted))  # Derivative w.r.t. m
        bd = -(2/n) * sum(y - y_predicted)        # Derivative w.r.t. b

        # Update parameters
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * bd

    # Plot the final regression line
    final_prediction = m_curr * x + b_curr
    plt.plot(x, final_prediction, color='blue', label='Final Regression Line')
    plt.xlabel("X values")
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)

scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(digits.data,digits.target):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
                                       digits.target[train_index], digits.target[test_index]
    scores_logistic.append(get_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X_train, X_test, y_train, y_test))
    scores_svm.append(get_score(SVC(gamma='auto'), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
```
